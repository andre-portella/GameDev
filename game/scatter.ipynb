```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
```

```python

plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.scatter(x=x, y=y, color='b', s=50, alpha=0.7, edgecolors='k', label='Pontos')  # Grfico de disperso
plt.plot(x, y, color='r', linestyle='-', linewidth=2, label='Linha')  # Adiciona a linha conectando os pontos
plt.xlabel('Tempo (s)', fontsize=14)  # Define o rtulo do eixo x com tamanho de fonte
plt.ylabel('Velocidade (m/s)', fontsize=14)  # Define o rtulo do eixo y com tamanho de fonte
plt.title('Grfico de Disperso de Tempo vs. Velocidade', fontsize=16)  # Adiciona um ttulo ao grfico
plt.grid(True)  # Adiciona uma grade ao grfico para melhor visualizao
plt.legend()  # Adiciona uma legenda ao grfico
plt.show()  # Exibe o grfico
```

```python
# Supondo que 'x' e 'y' so suas variveis com os dados
# x = [seus dados para o eixo x]
# y = [seus dados para o eixo y]

# Ajuste de uma curva polinomial de grau 2 (parbola)
coeficientes = np.polyfit(x, y, 2)
polinomio = np.poly1d(coeficientes)

# Gera valores de x para a linha parablica
x_parabola = np.linspace(min(x), max(x), 500)
y_parabola = polinomio(x_parabola)

plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.scatter(x, y, color='blue', s=50, alpha=0.7, edgecolors='k', label='Pontos')  # Grfico de disperso
plt.plot(x_parabola, y_parabola, color='r', linestyle='--', linewidth=2, label='Parbola')  # Adiciona a linha parablica
plt.xlabel('Tempo (s)', fontsize=14)  # Define o rtulo do eixo x com tamanho de fonte
plt.ylabel('Velocidade (m/s)', fontsize=14)  # Define o rtulo do eixo y com tamanho de fonte
plt.title('Grfico Tempo vs Velocidade', fontsize=16)  # Adiciona um ttulo ao grfico
plt.grid(True)  # Adiciona uma grade ao grfico para melhor visualizao
plt.legend()  # Adiciona uma legenda ao grfico
plt.show()  # Exibe o grfico
```
